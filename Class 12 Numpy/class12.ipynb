## Scalar value

```python
import numpy as np

a : np.ndarray = np.array(1000) # object to store

display(a) # prints
display(a.shape) # prints the shape of the object
display(a.dtype) # prints the dtype of the object
display(a.ndim) # prints the number of the object
display(a.size) # prints the size of the object
display(a.itemsize) # prints the itemsize of the object
```

## Vector Type

```python
import numpy as np

a : np.ndarray = np.array([1,2,3,4]) # object to store [1,2,3,4] = vector

display(a) # prints
display(a.shape) # prints the shape of the object
display(a.dtype) # prints the dtype of the object
display(a.ndim) # prints the number of the object
display(a.size) # prints the size of the object
display(a.itemsize) # prints the itemsize of the object
```

## Numpy With NDArray Typing Support

```python
from nptyping import NDArray, UInt64, Shape

data : NDArray[Shape["10"], UInt64] = np.arange(1 ,11)

data
```

## Numpy Advanced Operations vs List Operations

```python
ndata : NDArray[Shape["20"], UInt64] = np.arange (1 ,21)

print(ndata)
print(ndata[5:11])
ndata[5:11] = 1000
print(ndata)
```

# Numpy 1d : (dimension) Functions

```python
from nptyping import Bool

state_bank : NDArray[Shape["10"], UInt64] = np.array([1,7,8,10])
# [1, 7, 8, 10]
# [True, False, False, True]
# variable [True, False, False, True]
select : NDArray[Shape["10"], Bool] = np.array([True, False, False, True])

state_bank[select]
```

```python
state_bank : NDArray[Shape["10"], UInt64] = np.array([1,7,8,10])
state_bank % 2 == 0
```

```python
state_bank : NDArray[Shape["10"], UInt64] = np.array([1,7,8,10])
state_bank [state_bank % 2 == 0]
```

```python
state_bank : NDArray[Shape["10"], UInt64] = np.array([1,7,8,10])

UBL_bank : NDArray[Shape["100"], UInt64] = np.random.randint(1, 100, 20)

print(state_bank)
print(UBL_bank)
```

```python
display(state_bank)
display(UBL_bank)

np.in1d(state_bank, UBL_bank)
```

```python
np.intersect1d(state_bank, UBL_bank)
```

# Create Numpy Array

```python
a : NDArray[Shape["Size, Size"], UInt64] = np.array([[1,2,3],[4,5,6]])

print(a)

a : NDArray[Shape["Size, Size"], UInt64] = np.array([[1,2],[4,5]])

print(a)

a : NDArray[Shape["Size, Size"], UInt64] = np.array([["A"],["B"]])

print(a)
```

```python
a : NDArray[Shape["*, *"], UInt64] = np.array([[1,2,3],[4,5,6]])

print(a)

a : NDArray[Shape["*, *"], UInt64] = np.array([[1,2],[4,5]])

print(a)

a : NDArray[Shape["*, *"], UInt64] = np.array([["A"],["B"]])

print(a)
```

## Create any dimension array

```python
a : NDArray[Shape["Size"], UInt64] = np.arange(1,5)

print(a)

a : NDArray[Shape["Size"], UInt64] = np.arange(1,10)

print(a)
```

```python
a : NDArray[Shape["Size, Size"], UInt64] = np.arange(3*3).reshape(3,3)

print(a)

a : NDArray[Shape["Size"], UInt64] = np.arange(5*3).reshape(5,3)

print(a)
```
