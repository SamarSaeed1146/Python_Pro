```python
total = 10000

# Creating the arrays
s1 = np.random.randint(80, 100, (total, 5))
ss1 = pd.DataFrame(s1, columns=['s1','s2','s3','s4','s5'])

s2 = np.random.randint(70, 79, (total, 5))
ss2 = pd.DataFrame(s2, columns=['s1','s2','s3','s4','s5'])

s3 = np.random.randint(60, 69, (total, 5))
ss3 = pd.DataFrame(s3, columns=['s1','s2','s3','s4','s5'])

s4 = np.random.randint(50, 59, (total, 5))
ss4 = pd.DataFrame(s4, columns=['s1','s2','s3','s4','s5'])

s5 = np.random.randint(40, 49, (total, 5))
ss5 = pd.DataFrame(s5, columns=['s1','s2','s3','s4','s5'])

s6 = np.random.randint(33, 39, (total, 5))
ss6 = pd.DataFrame(s6, columns=['s1','s2','s3','s4','s5'])

s7 = np.random.randint(0, 32, (total, 5))
ss7 = pd.DataFrame(s7, columns=['s1','s2','s3','s4','s5'])
```

```python
df : pd.DataFrame = pd.concat([ss1, ss2, ss3, ss4, ss5, ss6, ss7])

display(df.info())
print("=====+=====+=====+=====+=====+=====+=====+=====+")
display(df.describe())
print("=====+=====+=====+=====+=====+=====+=====+=====+")
display(df.head())
print("=====+=====+=====+=====+=====+=====+=====+=====+")
display(df.tail())
```
